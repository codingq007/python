```python
class Book:
    def __init__(self, title, code, author, publisher):
        self.title = title        
        self.code = code
        self.author = author
        self.publisher = publisher
        
    def book_info(self):
        print('도서코드 : ', self.code)
        print('도서명 : ', self.title)
        print('저 자 : ', self.author)
        print('출판사 : ', self.publisher)

def menu_display():
    print('--------------------')
    print('1. 도서 정보 입력')
    print('2. 도서 정보 출력')
    print('3. 도서 삭제')
    print('4. 종료')
    print('--------------------')
    
    menu = input('메뉴 선택 : ')
    return int(menu)        
        
# 도서 정보를 입력        
def set_book():# input 이용해서 입력받은 값을 출력하는 함수
        code = input('도서코드 : ')
        title = input('도서명 : ')
        author = input('저자 :')
        publisher = input('출판사 : ')
        
        print()
#         print(code, title, author, publisher)       
        book = Book(title, code, author, publisher)
        return book

# 도서 정보 출력
def print_book_info(book_list):
    for book in book_list:
        book.book_info()
    
def delete_book(book_list, title):
    for idx, book in enumerate(book_list):
        if book.title == title:
            del book_list[idx]

            
def repo_book(book_list):
    with open('./book_db.txt', 'w', encoding='utf8') as f:
        for book in book_list:
            f.write(book.title + '\n')
            f.write(book.code + '\n')
            f.write(book.author + '\n')
            f.write(book.publisher + '\n')

            
def load_book(book_list):
    with open('./book_db.txt', 'r', encoding='utf8') as f:
        lines = f.readlines()
        print(lines)
        num = len(lines) / 4 #book의 수 파악
        num = int(num)
        
        for i in range(num):
            title = lines[4*i]
            code = lines[4*i+1]
            author = lines[4*i+2]
            publisher = lines[4*i+3]
            
            book = Book(title, code, author, publisher)
            book_list.append(book)
            
        

def run():    
    book_list = []
    load_book(book_list)
    
    while 1:
        menu = menu_display()
        
        if menu == 1:
            book = set_book()
            book_list.append(book)
        elif menu == 2:
            print_book_info(book_list)

        elif menu == 3:
            title = input('삭제할 도서명을 입력하세요 : ')
            delete_book(book_list, title)
        elif menu == 4:
            repo_book(book_list)
            break


if __name__ == '__main__':    
    run()    
```
